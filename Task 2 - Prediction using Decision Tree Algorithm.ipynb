### Create the Decision Tree classifier and visualize it graphically. The purpose is if we feed any new data to this classifier, it would be able to  predict the right class accordingly.

```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = pd.read_csv("Iris.csv")
```

```python
iris.head()
```

```python
iris.info()
```

```python
iris.drop('Id',axis=1,inplace=True) 
#dropping the Id column as it is unecessary, 
#axis=1 specifies that it should be column wise, 
#inplace=1 means the changes should be reflected into the dataframe
```

### Splitting the data into train and test data

```python
from sklearn.model_selection import train_test_split
```

```python
train, test = train_test_split(iris, test_size = 0.3)# in this our main data is split into train and test
# the attribute test_size=0.3 splits the data into 70% and 30% ratio. train=70% and test=30%
print(train.shape)
print(test.shape)
```

```python
train_X = train[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]# taking the training data features
train_y = train.Species  #output of our training data
test_X = test[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']] # taking test data features
test_y = test.Species   #output value of test data
```

### Decision Tree 

```python
from sklearn import metrics
from sklearn.tree import DecisionTreeClassifier
```

```python
model=DecisionTreeClassifier()
model.fit(train_X,train_y)
prediction=model.predict(test_X)
print('The accuracy of the Decision Tree is',metrics.accuracy_score(prediction,test_y))
```

### Graphical Representation of Decision Tree

```python
from sklearn.tree import plot_tree

feature = iris.columns
classes = iris['Species'].unique().tolist()
plt.figure(figsize=(15,10))
plot_tree(model, feature_names=feature, class_names=classes, filled=True);
```
